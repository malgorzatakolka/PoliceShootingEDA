```python
# Let's make a dataframe out of those values
killed = pd.DataFrame({'race':['white', 'black', 'hispanic', 'others'], 
                      'num_killed':[3643, 1914, 1289, 142+111+20], 
                      'pop_percent':[white, black, hispanic, others]})
killed
```

```python
# Killed per race column
pd.set_option('display.float_format', '{:.2f}'.format)
killed['killed_per_million_per_year'] = killed['num_killed']*1000000/(killed['pop_percent']*population*7)
killed = killed.sort_values(by='killed_per_million_per_year', ascending=False)
killed
```
